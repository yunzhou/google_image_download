## get all scientific names

```python
succlent_scienticif_names=[]
with open('Browse Succulents by Scientific Name _ World of Succulents.html') as f:
    html_doc=f.read()
    soup = BeautifulSoup(html_doc, 'html.parser')
    ts=soup.find_all("div", attrs={'class':'tagindex'})
    for t in ts:
        links=t.find_all('li')
        for l in links:
            if 'class' in l.attrs and l['class']<>'morelink':
                pass
            else:
                s=l.a['title'].encode('ascii',errors='ignore')
                link=l.a['href']
                succlent_scienticif_names.append((s,link))
```
